```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets  
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
```

```python
# import some data to play with
df= pd.read_csv("C:/Users/khann/OneDrive/Desktop/excel r assignments/Company_Data.csv")
df.info()
```

```python

df.loc[df.Sales<=8, "Sales"]='Bad'
df.loc[df.Sales!='Bad', "Sales"]= 'Good'

df
```

```python
data_cleaned=pd.get_dummies(df,columns=['ShelveLoc','Urban','US'],drop_first=True)
data_cleaned
```

```python
#Complete fraud dataset
data=data_cleaned.iloc[:,1:12]
data
```

```python
target=data_cleaned.iloc[:,0]
target
```

```python
x=data
y=target
```

```python
x
```

```python
y
```

```python
# Splitting data into training and testing data set
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=40)
```

### Building Decision Tree Classifier using Entropy Criteria

```python
model = DecisionTreeClassifier(criterion = 'entropy',max_depth=3)
model.fit(x_train,y_train)
```

```python
model.get_n_leaves()
```

```python
preds = model.predict(x_test) # predicting on test data set 
pd.Series(preds).value_counts() # getting the count of each category 
```

```python
preds
```

```python
pd.crosstab(y_test,preds) # getting the 2 way table to understand the correct and wrong predictions 
```

```python
# Accuracy 
np.mean(preds==y_test)
```

```python
print(classification_report(preds,y_test))
```
